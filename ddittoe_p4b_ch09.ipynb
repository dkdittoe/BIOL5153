```python
# binning DNA sequences
#create 9 new folders 
#100 to 199, 200 to 299, etc

import os

#create new folders
for bin_lower in range(100,1000,100):
    bin_upper = bin_lower + 99
    bin_folder = str(bin_lower) + '__' + str(bin_upper)
    os.mkdir(bin_folder)
    
#variable to hold sequence number
seq_num = 1

#get .dna files
for files in os.listdir('.'):
    if files.endswith('.dna'):
        print ('sequences from ' + files)
        files_dna = open(files)
        
        #look at each line and calculate length of sequences
        for line in files_dna:
            dna = line.rstrip('\n')
            length = len(dna)
            print('dna sequence length ' + str(length))
            
            #bin the files
            for bin_lower in range(100,1000,100):
                bin_upper = bin_lower + 99
                if length >= bin_lower and length < bin_upper:
                    print('bin is ' + str(bin_lower) + ' to ' + str(bin_upper))
                    bin_folder = str(bin_lower) + '__' + str(bin_upper)
                    output_name = bin_folder + '/' + str(seq_num) + '.dna'
                    output = open(output_name, 'w')
                    output.write(dna)
                    output.close()
                    
                    seq_num = seq_num+1
                    
                    
                    
```

```python
# def is signature, split_dna is the function name, and dna/kmer_size is the parameter
def split_dna(dna, kmer_size):
    # kmers is the local variable
    kmers = []
    for start in range (0, len(dna)-(kmer_size-1),1):
        kmer = dna[start:start+kmer_size-1]
        kmers.append(kmer)
    return kmers

print(split_dna('ATGCTGACGTAAGCT', 4))
print(split_dna('GTACGTACGGACTGA', 5))
print(split_dna('ATCGGACTGACTGAC', 6))
```

```python
#kmer counting
import os
import sys

kmer_size = int(sys.argv[1])
count_cutoff = int(sys.argv[2])

def split_dna(dna, kmer_size):
    # kmers is the local variable
    kmers = []
    for start in range (0, len(dna)-(kmer_size-1),1):
        kmer = dna[start:start+kmer_size-1]
        kmers.append(kmer)
    return kmers

#empty dictionary to hold the counts
kmer_counts = {}

#files to correct file names
for files in os.listdir('.'):
    if files.endswith('.dna'):
        dna_file = open(files)
        #line by line
        for line in dna_file:
            dna = line.rstrip("\n")
            #keept track of total kmer number
            for kmer in split_dna(dna, kmer_size):
                #get is a part of the disctionary 
                current_count = kmer_counts.get(kmer, 0)
                new_count = current_count + 1
                kmer_counts[kmer] = new_count
                
#print kmers above cutoff
for kmer, count in kmer_counts.items():
    if count > count_cutoff:
        print(kmer + ' : ' + str(count))
        
#to run on terminal type -> python [filename.py] [length] [cutoff]
```
